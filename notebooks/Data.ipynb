```python
import pandas as pd

def normalize_with_prefix(df, col):
    """
    Flatten a dict-like column `col` and prefix all new columns with `col.`
    Returns a new dataframe.
    """
    if col not in df.columns:
        return df  # nothing to do

    # Turn non-dict values into {} so json_normalize doesn't crash
    series = df[col].apply(lambda x: x if isinstance(x, dict) else {})

    # Flatten
    nested = pd.json_normalize(series)

    # Add prefix: user.id, user.followers_count, ...
    nested.columns = [f"{col}.{c}" for c in nested.columns]

    # Drop original nested column and concat flattened ones
    df = pd.concat([df.drop(columns=[col]), nested], axis=1)

    return df
```

```python
import numpy as np
import pandas as pd

# --------- Helpers ---------
def is_null(x):
    return x is None or (isinstance(x, float) and np.isnan(x))

def safe_count_list(x):
    """
    - NaN / None  -> np.nan
    - []          -> 0
    - list[...]   -> len(list)
    - anything else -> 0
    """
    if is_null(x):
        return np.nan
    if isinstance(x, list):
        return len(x)
    return 0

def safe_display_range_len(x):
    """
    For [start, end] -> end - start
    """
    if is_null(x):
        return np.nan
    if isinstance(x, list) and len(x) == 2:
        try:
            return x[1] - x[0]
        except Exception:
            return np.nan
    return np.nan

# --------- 1) extended_entities ‚Üí count of media ---------
def count_extended_entities_media(x):
    """
    x is usually a dict with key 'media': [ ... ]
    """
    if is_null(x):
        return np.nan
    if isinstance(x, dict):
        media = x.get("media", [])
        if isinstance(media, list):
            return len(media)
        return 0
    return 0

# 1) extended_entities ‚Üí count of media
df["extended_entities.count"] = df["extended_entities"].apply(count_extended_entities_media)

# 2) extended_tweet.display_text_range ‚Üí length (still a numeric feature, but keep naming consistent)
df["extended_tweet.display_text_range.count"] = df["extended_tweet.display_text_range"].apply(
    safe_display_range_len
)

# 3) List-like entity fields ‚Üí counts
df["extended_tweet.entities.urls.count"] = df["extended_tweet.entities.urls"].apply(safe_count_list)

df["extended_tweet.entities.hashtags.count"] = df["extended_tweet.entities.hashtags"].apply(safe_count_list)

df["extended_tweet.entities.user_mentions.count"] = df["extended_tweet.entities.user_mentions"].apply(safe_count_list)

df["extended_tweet.entities.symbols.count"] = df["extended_tweet.entities.symbols"].apply(safe_count_list)

df["extended_tweet.extended_entities.media.count"] = df["extended_tweet.extended_entities.media"].apply(safe_count_list)

df["extended_tweet.entities.media.count"] = df["extended_tweet.entities.media"].apply(safe_count_list)


# --------- (Optional) Drop high-cardinality original columns ---------
cols_to_drop = [
    "extended_entities",
    "extended_tweet.display_text_range",
    "extended_tweet.entities.urls",
    "extended_tweet.entities.hashtags",
    "extended_tweet.entities.user_mentions",
    "extended_tweet.entities.symbols",
    "extended_tweet.extended_entities.media",
    "extended_tweet.entities.media",
]

# If you‚Äôre ready to drop them:
df = df.drop(columns=cols_to_drop)
```

```python
import pandas as pd

def is_empty_list_cell(x):
    # Case 1: real Python empty list
    if isinstance(x, list) and len(x) == 0:
        return True
    # Case 2: string that looks exactly like "[]"
    if isinstance(x, str) and x.strip() == "[]":
        return True
    return False

empty_list_percent = {}

n_rows = len(df)

for col in df.columns:
    count_empty = df[col].apply(is_empty_list_cell).sum()
    empty_list_percent[col] = (count_empty / n_rows) * 100 if n_rows > 0 else 0

empty_list_percent = pd.Series(empty_list_percent, name="empty_list_percent")



# Or sorted (highest % of [] first)
print(empty_list_percent.sort_values(ascending=False))
```

```python
import pandas as pd
import numpy as np
import math

def count_listlike(value):
    # 1) True missing values
    if value is None:
        return np.nan
    
    # NaN as float
    if isinstance(value, float) and math.isnan(value):
        return np.nan

    # 2) If it's already a list or numpy array: count elements
    if isinstance(value, (list, np.ndarray)):
        return len(value)  # [] -> 0, [..] -> count

    # 3) If it's a string like "[]", "[{...}]", etc.
    if isinstance(value, str):
        s = value.strip()
        if s == "" or s == "[]":
            return 0
        # could try json.loads(s), but we keep it simple:
        return 0

    # 4) Fallback: treat as 0
    return 0
```

```python
import pandas as pd
import numpy as np

def hex_to_rgb(h):
    """
    Convert a hex string like '1DA1F2' or '#1DA1F2' to (r, g, b).
    - Keep nulls as NaN.
    - If parsing fails ‚Üí (NaN, NaN, NaN)
    """
    if pd.isna(h):
        return (np.nan, np.nan, np.nan)
    
    s = str(h).strip().lstrip("#")  # remove spaces and leading '#'
    
    if len(s) != 6:
        return (np.nan, np.nan, np.nan)
    
    try:
        r = int(s[0:2], 16)
        g = int(s[2:4], 16)
        b = int(s[4:6], 16)
        return (r, g, b)
    except ValueError:
        return (np.nan, np.nan, np.nan)


color_cols = [
    "user.profile_background_color",
    "user.profile_link_color",
    "user.profile_sidebar_border_color",
    "user.profile_text_color",
    "user.profile_sidebar_fill_color",
]

for col in color_cols:
    if col in df.columns:
        # each value ‚Üí (r,g,b), then expand to 3 columns
        rgb = df[col].apply(hex_to_rgb).apply(pd.Series)
        rgb.columns = [f"{col}_r", f"{col}_g", f"{col}_b"]

        df = pd.concat([df, rgb], axis=1)

        print(f"Created {col}_r, {col}_g, {col}_b")
    else:
        print(f"Column {col} not found in df")
```

```python
import pandas as pd
import numpy as np

def extract_entities_features(ents):
    if not isinstance(ents, dict) or ents is None:
        return pd.Series({"n_urls": 0, "n_hashtags": 0, "n_mentions": 0})

    urls = ents.get("urls", []) or []
    hashtags = ents.get("hashtags", []) or []
    mentions = ents.get("user_mentions", []) or []

    return pd.Series({
        "n_urls": len(urls),
        "n_hashtags": len(hashtags),
        "n_mentions": len(mentions)
    })

df[["entities.n_urls", "entities.n_hashtags", "entities.n_mentions"]] = df["entities"].apply(extract_entities_features)

# then drop the big dict column:
df = df.drop(columns=["entities"])
```
